Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

In Bokeh, glyphs are visual markers that represent data points on a plot. They can be used to display various types of data, such as points, lines, bars, or patches. Bokeh provides a wide range of glyph types that you can use to customize your plots.

```python
from bokeh.plotting import figure, show

# Create a figure object
p = figure()

# Add a scatter plot using circles
p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=10, color='blue', alpha=0.5)

# Add a line plot
p.line([1, 2, 3, 4, 5], [8, 5, 3, 9, 1], line_width=2, color='red')

# Add a rectangle
p.rect([2.5], [4], width=0.8, height=1.5, color='green', alpha=0.3)

# Add a patch
p.patch([3, 4, 4, 3], [1, 1, 2, 2], fill_color='orange', line_color='black', alpha=0.7)

# Show the figure
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
from bokeh.plotting import figure , output_file, show
from bokeh.sampledata.iris import flowers

output_file('line.html')

p = figure(title = "line graph")
p.xaxis.axis_label = "x axis"
p.yaxis.axis_label = "y axis"
p.line(flowers ['petal_length'], flowers['petal_width'], legend = 'This is my legend')
show(p)
```
